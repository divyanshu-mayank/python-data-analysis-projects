## Filling all the null enteries

```python
data['Cylinders'].fillna(data['Cylinders'].mean(), inplace=True)
```

Checking the frequency of different type of 'Makes' of the data

```python
data['Make'].value_counts()
```

Checking all the enteries with 'Origin' as Asia or Europe.

```python
data[data['Origin'].isin(['Asia', 'Europe'])]
```

```python
data[(data['Origin'] == 'Asia') | (data['Origin'] == 'Europe')]
```

## Remove all the records where weight is above 4000

```python
data.drop(data[data['Weight'] > 4000].index, inplace=True)

# Alternate Code
# data[~(data['Weight']>4000)]
```

```python
data
```

## Increasing all the vaues of 'MPG_City' column by 3

```python
data['MPG_City'] = data['MPG_City'].apply(lambda x: x+3)

data['MPG_City'] = data['MPG_City'] + 3
```
